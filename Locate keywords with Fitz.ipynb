```python
import fitz  # PyMuPDF

# Define the path to your PDF file
file_path = 'Huge.pdf'

# Open the PDF file
pdf_document = fitz.open(file_path)

# Define the keyword to search for
keyword = "AI"

# Store the results
results = []

# Iterate through each page
for page_num in range(len(pdf_document)):
    page = pdf_document.load_page(page_num)
    text_instances = page.search_for(keyword)
    
    # Get the positions of the keyword
    for inst in text_instances:
        x0, y0, x1, y1 = inst
        # Get the text blocks that contain the keyword
        blocks = page.get_text("blocks")
        for block in blocks:
            bx0, by0, bx1, by1, block_text, *_ = block
            if keyword in block_text:
                
                # Calculate row number or any other required detail
                results.append((page_num + 1, bx0, by0, bx1, by1, keyword))

# Print the results
for result in results:
    page_num, x0, y0, x1, y1, word = result
    print(f"'{word}' found on page {page_num}, position ({x0}, {y0}, {x1}, {y1})")
```
