```python

# Code to extract text from a PDF
import fitz  # PyMuPDF

def extract_text_from_pdf(file_path):
    pdf_document = fitz.open(file_path)
    text = ""
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text += page.get_text("text")
    return text

# Ensure the file 'ResumeSample.pdf' exists in the current directory
resume_text = extract_text_from_pdf('ResumeSample.pdf')
#print(resume_text)
```

## Semantic matching

```python
# Required Libraries
import spacy
from fuzzywuzzy import fuzz, process
from collections import defaultdict


# Load spacy model
nlp = spacy.load('en_core_web_sm')

# Example dictionary of section headers and associated phrases
section_headers = {
    "Executive Summary": ["executive summary", "summary of qualifications", "professional summary"],
    "Core Competencies": ["core competencies", "skills and expertise", "core skills"],
    "Skills and Technologies": ["skills and technologies", "technical skills", "technology proficiencies"],
    "Professional Experience": ["professional experience", "work experience", "employment history"],
    "Education": ["education", "academic background", "educational qualifications"]
}

def extract_relevant_sections(text):
    # Split text into lines for easier processing
    lines = text.split('\n')
    
    # Dictionary to hold section content
    sections = {}
    current_section = None

    # Identify section headers using fuzzy matching
    for line in lines:
        line = line.strip()  # Remove leading and trailing whitespace
        if not line:
            continue  # Skip empty lines
        
        # Perform fuzzy matching to find the closest section header
        header, score = process.extractOne(line, section_headers.keys(), scorer=fuzz.partial_ratio)
        if score > 80:  # Threshold for fuzzy matching
            current_section = header
            if current_section not in sections:
                sections[current_section] = []
            continue
        
        # If the line is not a section header, add it to the current section
        if current_section:
            sections[current_section].append(line)

    # Combine lines into a single string for each section
    for header in sections:
        sections[header] = ' '.join(sections[header]).strip()
    
    return sections

sections = extract_relevant_sections(resume_text)

print("Extracted Sections:")
for section, content in sections.items():
    print(f"\nSection: {section}")
    print(content)
```
